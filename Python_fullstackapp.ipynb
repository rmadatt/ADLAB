```python
# Install graphviz if not already available
!apt-get install -y graphviz
!pip install graphviz

from graphviz import Digraph

# Create a directed graph
dot = Digraph(comment="Python Full Stack Development Roadmap")

# Define nodes (steps)
dot.node("1", "1. Planning & Requirements")
dot.node("2", "2. System Design")
dot.node("3", "3. Backend Development")
dot.node("4", "4. Frontend Development")
dot.node("5", "5. Integration & Testing")
dot.node("6", "6. Deployment")
dot.node("7", "7. Maintenance & Iteration")

# Connect nodes (workflow)
dot.edges(["12", "23", "34", "45", "56", "67"])

# Render the diagram
dot.render("fullstack_roadmap", format="png", cleanup=True)

# Display in Colab
from IPython.display import Image
Image(filename="fullstack_roadmap.png")
```
